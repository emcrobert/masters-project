```python
%reload_ext autoreload
%autoreload 2

from helpers import *
import matplotlib.pyplot as plt
import numpy as np
import pprint
from tensorflow.keras.layers.experimental.preprocessing import Rescaling

DATASET_ROOT_PATH = '../../data/HiltonOfFern_crop_field_training_cloud_free_available_area/'

edge_of_roi_files = [
    '/20181029/ROI2/S2/Patches/S1_992a54fd-c4ef-4f33-ab51-79efd8f7f63a_S2_36ac7212-ead8-4abf-854a-b1043365e7d0_1024_2304_256x256.tif',
    '/20181029/ROI2/S2/Patches/S1_992a54fd-c4ef-4f33-ab51-79efd8f7f63a_S2_36ac7212-ead8-4abf-854a-b1043365e7d0_2560_3072_256x256.tif',
    '/20181029/ROI2/S2/Patches/S1_992a54fd-c4ef-4f33-ab51-79efd8f7f63a_S2_36ac7212-ead8-4abf-854a-b1043365e7d0_2304_1280_256x256.tif',
    '/20181015/ROI1/S2/Patches/S1_531e42bd-eadb-495e-ae77-e3242f5ed70e_S2_4d4ec32a-afff-40a3-b8a3-f0a23320a93c_2560_2304_256x256.tif',
    '/20190513/ROI1/S2/Patches/S1_e316e53f-5e52-486c-a5d5-e82e63d0c3bc_S2_780cee33-2b59-4171-8f6b-be8f221fbfca_1792_512_256x256.tif',
    '/20190225/ROI1/S2/Patches/S1_150c763c-0ee6-4611-a6ce-d9cc89f3420b_S2_d5bb8ed3-4ad7-4373-afd0-cfbfb5523da7_2048_3072_256x256.tif',
    '20180827/ROI1/S2/Patches/S1_f78df197-5814-44ea-bfff-351e5e1ddb50_S2_722cd3b3-d490-4d4d-8de5-891e19d25d95_0_2048_256x256.tif'
]

# have multiple bands with different cloud information (13 - opaque clouds, 14 cirrus clouds, 15 cloud shadow, 16 medium prob cloud,
# 17 high prob cloud, 18 thin cirrus). I just use 17-18 which can identify images with most obvious clouds
def get_cloud_mask_info(file_path):
    if type(file_path) is not str:
        file_path = compat.as_text(file_path.numpy()) 
        
    sentinel_data =  rasterio.open(DATASET_ROOT_PATH + file_path)
    
    test = sentinel_data.read(0)
    for cur_band in range(13, 19):
        cur_mask = sentinel_data.read(cur_band)
        print("cloud mask for band " + str(cur_band) + " contains " + str(cur_mask.size) + " values " + str(np.count_nonzero(cur_mask)) + " of which are non-zero" )
    

rescaling_layer = Rescaling(scale = 1./1811.0)

img_counter = 1
for cur_image in edge_of_roi_files:
    print("loading edge of roi file " + cur_image)
    sen2 = convert_sen2_product_to_rgb_tensor(DATASET_ROOT_PATH + cur_image, 1811.0, "clip")
    sen2 = tf.reshape(sen2, [256, 256, 3])
    sen2 = rescaling_layer(sen2)
    
    plt.figure(img_counter)
    plt.title("edge of roi file " + cur_image)
    plt.imshow(sen2)
    img_counter += 1
    
cloud_shadow_file = '/20180910/ROI1/S2/Patches/S1_af77da2e-fcc7-4c9f-9d70-4fc1a06c0ffc_S2_153d0c59-4118-4a16-b459-a2432b57b575_256_2560_256x256.tif'

print("loading rgb values for image with cloud shadow")
sen2 = convert_sen2_product_to_rgb_tensor(DATASET_ROOT_PATH + cloud_shadow_file, 1811.0, "clip")
print("loading cloud shadow mask")
cloud_shadow_mask = convert_sen2_product_to_cloud_shadow_mask_tensor(DATASET_ROOT_PATH + cloud_shadow_file)
total_mask_values = tf.cast(tf.size(cloud_shadow_mask), dtype=tf.float32)
cloud_vals = tf.math.count_nonzero(cloud_shadow_mask)
print("cloud shadow mask has " + str(cloud_vals) + " non zero values and " + str(cloud_vals) + " values in total")

sen2 = tf.reshape(sen2, [256, 256, 3])
sen2 = rescaling_layer(sen2)

plt.figure(img_counter)
plt.title("image in shadow " + cur_image)
plt.imshow(sen2)
img_counter += 1
    
cloud_file = '/20181008/ROI1/S2/Patches/S1_c003e1d5-2306-47b5-9497-659a90562f2d_S2_46edee3a-9f5d-4912-8e55-d9b684f1166b_768_1536_256x256.tif'
print("loading rgb values for image with cloud")
sen2 = convert_sen2_product_to_rgb_tensor(DATASET_ROOT_PATH + cloud_file, 1811.0, "clip")
sen2 = tf.reshape(sen2, [256, 256, 3])
sen2 = rescaling_layer(sen2)
get_cloud_mask_info(cloud_file)
plt.figure(img_counter)
plt.title("image with cloud " + cur_image)
plt.imshow(sen2)
```
